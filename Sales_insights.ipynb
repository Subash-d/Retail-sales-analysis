```python
#Importing required packages
import pandas as pd
```

```python
#Importing the file

df = pd.read_csv("SQL - Retail Sales Analysis_utf .csv")
df.head(5)
```

```python
df = df.drop_duplicates() #To remove duplicate rows. if availble
df
```

```python
df.isnull().sum()
```

```python
df[df.isnull(). any(axis = 1)]

#df.count()   #To see how many rows with null values
```

```python
#Filling the null values by apllying the mean

df['age'] = df['age'].fillna(int(round(df['age'].mean())))

df['total_sale'] = df['total_sale'].fillna(round(df['total_sale'].mean()))

df['quantiy'] = df['quantiy'].fillna(round(df['quantiy'].mean()))

df['price_per_unit'] = df['price_per_unit'].fillna(df['price_per_unit'].mean())

df['cogs'] = df['cogs'].fillna(df['cogs'].mean())
```

```python
#Null values were handeled efficiently
df.isnull().sum() 
```

```python
df.to_csv("Sales_insights.csv", index= False)
```
